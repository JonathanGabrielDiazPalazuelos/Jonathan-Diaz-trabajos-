1. $f(x)=x+\frac{1}{x^2}$

**Solución:**

Lo primero que haré es darle valor a la x, para despues definir  $f(x)$ como la función que es   $x+\frac{1}{x^2}$


Despues voy a derivar la $f(x)$, para luego resolverla y luego la volveré a derivar.

Luego haremos el uso de if, elif y else para poder saber si e sun máximo, mínimo o un punto de inflexión.

Finalmente, haré una grafica marcando los puntos críticos, para demostrar que es cada punto crítico

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

x=sym.symbols("x")

def f(x):
  return x+(1/x**2)

df=f(x).diff(x)
c=sym.solve(df,x)
d2f=df.diff(x)

def maxMin(point):
  sdd=d2f.subs(x,point)
  if sdd>0:
    return "mínimo"
  elif sdd<0:
    return "máximo"
  else:
    return"punto de inflexión"

ppNum=[float(p) for p in c if p.is_real]

for point in ppNum:
  print(f"El punto crítico en {point} es un {maxMin(point)} global")

xNum=np.linspace(0,4,100)
yNum=[f(x) for x in xNum]

plt.plot(xNum,yNum,label="f(x)")
for point in ppNum:
  plt.plot(point,f(point),"*",ms=10)
plt.title("Gráfico de f(x)=x+(1/x**2)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

4. $f(x)=\frac{15x}{2}\sqrt{\frac{x}{x-30}}$

**Solución:**

Lo primero que haré es darle valor a la x, para despues definir  $f(x)$ como la función que es $5+x+\frac{4}{x}$


Despues voy a derivar la $f(x)$, para luego resolverla y luego la volveré a derivar.

Luego haremos el uso de if, elif y else para poder saber si e sun máximo, mínimo o un punto de inflexión.

Finalmente, haré una grafica marcando los puntos críticos, para demostrar que es cada punto crítico

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

x=sym.symbols("x")

def f(x):
  return ((15*x)/2)*((x/(x-30))**(1/2))

df=f(x).diff(x)
c=sym.solve(df,x)
d2f=df.diff(x)

def maxMin(q):
  sdd=d2f.subs(x,q)
  if np.isnan(float(sdd)):
    return "punto no determinado"
  if sdd>0:
    return "mínimo"
  elif sdd<0:
    return "máximo"
  else:
    return"punto de inflexión"

ppNum=[float(p) for p in c if p.is_real]

for q in ppNum:
  print(f"El punto crítico en {q} es un {maxMin(q)}")

xNum=np.linspace(-30,50,150)
yNum=[f(x) for x in xNum]

plt.plot(xNum,yNum)
for point in ppNum:
  plt.plot(point,f(point),"*",ms=10)
plt.title("Gráfico de f(x)=(15*x)/2)*(sym.sqrt(x)/(x-30)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
